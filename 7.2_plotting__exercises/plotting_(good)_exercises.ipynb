```python
from matplotlib import pyplot as plt
import numpy as np
fig, ax = plt.subplots(nrows=2, ncols=1)
x = np.linspace(-np.pi, np.pi, num = 100)
cos_x = np.cos(x)
sin_x = np.sin(x)
ax[0].plot(x, cos_x, label='cos_x')
ax[1].plot(x, sin_x, label='sin_x')
ax[0].legend()
ax[1].legend()
```
